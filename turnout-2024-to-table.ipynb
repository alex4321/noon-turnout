```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/turnout_voters_count_day3.tsv", delimiter="\t")
df
```

```python
def convert(df):
    records = []
    columns = [
        "turnout_15",
        "turnout_20",
        "turnout_2_15",
        "turnout_2_20",
        "turnout_3_10",
        "turnout_3_12",
        "turnout_3_15",
        "turnout_3_18",
    ]
    times = [
        pd.to_datetime("15.03.2024 15:00:00", dayfirst=True),
        pd.to_datetime("15.03.2024 20:00:00", dayfirst=True),
        pd.to_datetime("16.03.2024 15:00:00", dayfirst=True),
        pd.to_datetime("16.03.2024 20:00:00", dayfirst=True),
        pd.to_datetime("17.03.2024 10:00:00", dayfirst=True),
        pd.to_datetime("17.03.2024 12:00:00", dayfirst=True),
        pd.to_datetime("17.03.2024 15:00:00", dayfirst=True),
        pd.to_datetime("17.03.2024 18:00:00", dayfirst=True),
    ]
    for _, row in df.iterrows():
        for column, time in zip(columns, times):
            record = {}
            record["region"] = row["region"]
            record["tik"] = row["tik"]
            record["uik"] = row["uik"]
            record["time"] = time
            record["turnout_percent"] = row[f"{column}_pct"]
            records.append(record)
    return pd.DataFrame.from_records(records)
        
```

```python
df_converted = convert(df)
df_converted
```

```python
df_converted.to_csv("data/turnout-converted.csv", index=False)
```
